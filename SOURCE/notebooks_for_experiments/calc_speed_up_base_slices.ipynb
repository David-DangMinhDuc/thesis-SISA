```python
import pandas as pd
import cv2
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
file_lst_orl = ['/content/thesis-SISA/SOURCE/results/orl/report_1_shards_1_slices_30_epochs.csv',
'/content/thesis-SISA/SOURCE/results/orl/report_5_shards_1_slices_30_epochs.csv',
'/content/thesis-SISA/SOURCE/results/orl/report_5_shards_2_slices_30_epochs.csv',
'/content/thesis-SISA/SOURCE/results/orl/report_5_shards_3_slices_30_epochs.csv']
```

```python
file_lst_ar = ['/content/thesis-SISA/SOURCE/results/ar/report_1_shards_1_slices_10_epochs.csv',
'/content/thesis-SISA/SOURCE/results/ar/report_5_shards_1_slices_10_epochs.csv',
'/content/thesis-SISA/SOURCE/results/ar/report_5_shards_2_slices_10_epochs.csv']
```

```python
print("AR Face Database")
df_1_1 = pd.read_csv(file_lst_ar[0])
print(f"Case shard = 1, slice = 1")
print(f"Time: {round(df_1_1['retraining_time'].mean(),2)}")
print(f"Min speed up: 1")
print(f"Max speed up: 1")
print(f"Average speed up: 1")
print()

for ar in [file_lst_ar[1], file_lst_ar[2]]:
  df_u = pd.read_csv(ar)
  info = np.array(re.findall('\d+', ar), dtype=int)
  print(f"Case shard = {info[0]}, slice = {info[1]}")
  print(f"Time: {round(df_u['retraining_time'].mean(),2)}")
  print(f"Min speed up: {round((df_1_1['retraining_time'].values / df_u['retraining_time'].values).min(),2)}, with a number of request are {df_u['nb_requests'].values[np.argmin((df_1_1['retraining_time'].values / df_u['retraining_time'].values))]}")
  print(f"Max speed up: {round((df_1_1['retraining_time'].values / df_u['retraining_time'].values).max(),2)}, with a number of request are {df_u['nb_requests'].values[np.argmax((df_1_1['retraining_time'].values / df_u['retraining_time'].values))]}")
  print(f"Average speed up: {round((df_1_1['retraining_time'].values / df_u['retraining_time'].values).mean(),2)}")
  print()
```
